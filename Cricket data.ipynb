```python
import pandas as pd
import json
```

**Match Result**

```python
with open("C:\\Users\\Kunal\\Downloads\\Project\\t20 Json file\\t20_json_files\\t20_wc_match_results.json") as f:
    data=json.load(f)
#data                       
data[0]['matchSummary']  # these data can be arranged in well manneer
df_match=pd.DataFrame(data[0]['matchSummary'])  
df_match.head()
```

```python
df_match.shape          # how many Rows and Colum in File 
```

```python
df_match.rename({'scorecard':'match_id'},axis=1,inplace=True)
df_match.head()
```

```python
df_match.to_csv("C:\\Users\\Kunal\\Downloads\\self\\match_summary.csv",index=False)
```

```python
match_ids_dict= {}

for index,row in df_match.iterrows():
    key1=row['team1'] + ' Vs ' + row['team2']
    key2=row['team2'] + ' Vs ' + row['team1']
    
    match_ids_dict[key1]=row["match_id"]
    match_ids_dict[key2]=row["match_id"]
    
match_ids_dict
```

**Batting Summary**

```python
with open("C:\\Users\\Kunal\\Downloads\\Project\\t20 Json file\\t20_json_files\\t20_wc_batting_summary.json") as p:
    data=json.load(p)
 #In onteher way to append the file above manner   
    all_record=[]
    for rec in data:
        all_record.extend(rec['battingSummary'])
        
df_batting=pd.DataFrame(all_record)
df_batting.tail()
#df_match1=pd.DataFrame(data[0]['battingSummary'])
#df_match1.head()

```

```python
df_batting["out/not_out"]=df_batting.dismissal.apply(lambda x:"out" if len(x) > 0 else "not_out")
df_batting.head(11)
```

```python
df_batting.drop(columns=["dismissal"],inplace=True)
df_batting.head()
```

```python
df_batting['batsmanName']=df_batting['batsmanName'].apply(lambda x: x.replace('â€',''))
df_batting['batsmanName']=df_batting['batsmanName'].apply(lambda x: x.replace('\a0',''))
df_batting
```

```python
df_batting["match_id"]=df_batting["match"].map(match_ids_dict)
df_batting.head()
```

```python
df_batting.to_csv("C:\\Users\\Kunal\\Downloads\\self\\facts_batting_summary.csv",index=False)
```
